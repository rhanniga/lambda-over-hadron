```python
import ROOT as rt
```

```python
input_file = rt.TFile("AnalysisResults.root")
input_list = input_file.Get("h-lambda")
input_file.Close()
```

```python
c = rt.TCanvas()
```

```python
lambda_dist = input_list.FindObject("fDphiHLambdaEff")
lambda_dist_highestPt = input_list.FindObject("fDphiHLambdaEff_highestPt")
```

```python
lambda_dist.GetAxis(0).SetRangeUser(4, 8-0.00001)
lambda_dist_highestPt.GetAxis(0).SetRangeUser(4, 8-0.00001)
lambda_dist.GetAxis(1).SetRangeUser(2, 4-0.00001)
lambda_dist_highestPt.GetAxis(1).SetRangeUser(2, 4-0.00001)
```

```python
a = lambda_dist.Projection(4)
b = lambda_dist_highestPt.Projection(4)
a.SetTitle("Inv Mass Dist (blue = single trigger, red = all 4-8)")
a.SetLineColor(rt.kRed)
a.SetLineWidth(3)
```

```python
a.Draw()
b.Draw("SAME")
c.Draw()
```
