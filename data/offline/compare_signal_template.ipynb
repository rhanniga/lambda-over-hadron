```python
import ROOT as rt
```

```python
c = rt.TCanvas("c","c",800,600)
```

```python
SIG_MIN = 1.10
SIG_MAX = 1.14
```

```python
template_infile = rt.TFile("mc_templates.root", "READ")
lambda_mass_template = template_infile.Get("lambda_mass_dist_nobg")

data_infile = rt.TFile("output/res_avg6_sideband_subtraction_rsb_116_118_sig_11_114_trig_40_80_assoc_20_40_delta_eta_14.root", "READ")
residual_0_20 = data_infile.Get("residual_0_20")
residual_20_50 = data_infile.Get("residual_20_50")
residual_50_80 = data_infile.Get("residual_50_80")
```

```python
lambda_mass_template.Rebin(2)
```

```python
left_bin = lambda_mass_template.FindBin(SIG_MIN)
right_bin = lambda_mass_template.FindBin(SIG_MAX)
lambda_mass_template.Scale(residual_0_20.Integral(left_bin, right_bin)/lambda_mass_template.Integral(left_bin, right_bin))
```

```python
residual_0_20.GetXaxis().SetRangeUser(1.06, 1.18)
residual_0_20.Draw()
lambda_mass_template.Draw("SAME")
c.Draw()
```

```python

lambda_mass_template.Scale(residual_20_50.Integral(left_bin, right_bin)/lambda_mass_template.Integral(left_bin, right_bin))
```

```python
residual_20_50.GetXaxis().SetRangeUser(1.06, 1.18)
residual_20_50.Draw()
lambda_mass_template.Draw("SAME")
c.Draw()
```

```python
lambda_mass_template.Scale(residual_50_80.Integral(left_bin, right_bin)/lambda_mass_template.Integral(left_bin, right_bin))
```

```python
residual_50_80.GetXaxis().SetRangeUser(1.06, 1.18)
residual_50_80.Draw()
lambda_mass_template.Draw("SAME")
c.Draw()
```
