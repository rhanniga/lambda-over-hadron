```python
import ROOT as rt
rt.gStyle.SetOptStat(0)
```

```python
c = rt.TCanvas("c","c",800,600)
c.SetRightMargin(0.15)
c.SetLeftMargin(0.10)
c.SetTopMargin(0.05)
```

```python
ifile = rt.TFile.Open("~/AnalysisResults_partial.root")
# ifile = rt.TFile.Open("~/tmp1.root")
ilist = ifile.Get("h-lambda")
```

```python
fTOFnSigmaProton = ilist.FindObject("fTOFnSigmaProton")
fTOFnSigmaPion = ilist.FindObject("fTOFnSigmaPion")

fTPCnSigmaProton = ilist.FindObject("fTPCnSigmaProton")
fTPCnSigmaPion = ilist.FindObject("fTPCnSigmaPion")

fTPCvTOFnSigmaProton = ilist.FindObject("fTOFvTPCnSigmaProton")
fTPCvTOFnSigmaPion = ilist.FindObject("fTOFvTPCnSigmaPion")
```

```python
fTOFnSigmaProton.SetTitle("")
fTOFnSigmaProton.GetXaxis().SetTitle("#it{p}_{T} (GeV/c)")
fTOFnSigmaProton.GetXaxis().SetRangeUser(0.25, 10)
fTOFnSigmaProton.GetYaxis().SetTitle("n#sigma_{TOF, p, global}")
fTOFnSigmaProton.Draw("COLZ")

c.SetLogx()
c.Draw()
c.SaveAs("figures/nsigma_tof_proton_global.pdf")
```

```python
fTOFnSigmaPion.SetTitle("")
fTOFnSigmaPion.GetXaxis().SetTitle("#it{p}_{T} (GeV/c)")
fTOFnSigmaPion.GetXaxis().SetRangeUser(0.25, 10)
fTOFnSigmaPion.GetYaxis().SetTitle("n#sigma_{TOF, #pi, global}")
fTOFnSigmaPion.Draw("COLZ")
c.Draw()
c.SaveAs("figures/nsigma_tof_pion_global.pdf")
```

```python
fTPCnSigmaProton.SetTitle("")
fTPCnSigmaProton.GetXaxis().SetTitle("#it{p}_{T} (GeV/c)")
fTPCnSigmaProton.GetXaxis().SetRangeUser(0.25, 10)
fTPCnSigmaProton.GetYaxis().SetTitle("n#sigma_{TPC, p, global}")
fTPCnSigmaProton.Draw("COLZ")
c.Draw()
c.SaveAs("figures/nsigma_tpc_proton_global.pdf")
```

```python
fTPCnSigmaPion.SetTitle("")
fTPCnSigmaPion.GetXaxis().SetTitle("#it{p}_{T} (GeV/c)")
fTPCnSigmaPion.GetXaxis().SetRangeUser(0.25, 10)
fTPCnSigmaPion.GetYaxis().SetTitle("n#sigma_{TPC, #pi, global}")
fTPCnSigmaPion.Draw("COLZ")
c.Draw()
c.SaveAs("figures/nsigma_tpc_pion_global.pdf")
```

```python
fTPCvTOFnSigmaProton.SetTitle("")
fTPCvTOFnSigmaProton.GetYaxis().SetTitle("n#sigma_{TOF, p, global}")
fTPCvTOFnSigmaProton.GetXaxis().SetTitle("n#sigma_{TPC, p, global}")
fTPCvTOFnSigmaProton.Draw("COLZ")
c.Draw()
c.SaveAs("figures/nsigma_tof_v_tpc_proton.pdf")
```
