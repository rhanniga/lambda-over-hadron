```python
import ROOT as rt
rt.gStyle.SetOptStat(0)
```

```python
c = rt.TCanvas("c","c",800,600)
c.SetRightMargin(0.15)
c.SetLeftMargin(0.10)
c.SetTopMargin(0.05)
```

```python
ifile = rt.TFile.Open("../online/v0/AnalysisResults.root")
ilist = ifile.Get("h-lambda")
```

```python
fTOFnSigmaProton = ilist.FindObject("fTOFnSigmaProton")
fTOFnSigmaPion = ilist.FindObject("fTOFnSigmaPion")

fTPCnSigmaProton = ilist.FindObject("fTPCnSigmaProton")
fTPCnSigmaPion = ilist.FindObject("fTPCnSigmaPion")

fTPCvTOFnSigmaProton = ilist.FindObject("fTPCvTOFnSigmaProton")
fTPCvTOFnSigmaPion = ilist.FindObject("fTPCvTOFnSigmaPion")
```

```python
fTOFnSigmaProton.SetTitle("")
fTOFnSigmaProton.GetXaxis().SetTitle("#it{p}_{T} (GeV/c)")
fTOFnSigmaProton.GetYaxis().SetTitle("n#sigma_{TOF, p}")
fTOFnSigmaProton.Draw("COLZ")
c.Draw()
c.SaveAs("figures/nsigma_tof_proton.pdf")
```

```python
fTOFnSigmaPion.SetTitle("")
fTOFnSigmaPion.GetXaxis().SetTitle("#it{p}_{T} (GeV/c)")
fTOFnSigmaPion.GetYaxis().SetTitle("n#sigma_{TOF, #pi}")
fTOFnSigmaPion.Draw("COLZ")
c.Draw()
c.SaveAs("figures/nsigma_tof_pion.pdf")
```

```python
fTPCnSigmaProton.SetTitle("")
fTPCnSigmaProton.GetXaxis().SetTitle("#it{p}_{T} (GeV/c)")
fTPCnSigmaProton.GetYaxis().SetTitle("n#sigma_{TPC, p}")
fTPCnSigmaProton.Draw("COLZ")
c.Draw()
c.SaveAs("figures/nsigma_tpc_proton.pdf")
```

```python
fTPCnSigmaPion.SetTitle("")
fTPCnSigmaPion.GetXaxis().SetTitle("#it{p}_{T} (GeV/c)")
fTPCnSigmaPion.GetYaxis().SetTitle("n#sigma_{TPC, #pi}")
fTPCnSigmaPion.Draw("COLZ")
c.Draw()
c.SaveAs("figures/nsigma_tpc_pion.pdf")
```

```python
fTPCvTOFnSigmaProton.SetTitle("")
fTPCvTOFnSigmaProton.GetYaxis().SetTitle("n#sigma_{TOF, p}")
fTPCvTOFnSigmaProton.GetXaxis().SetTitle("n#sigma_{TPC, p}")
fTPCvTOFnSigmaProton.Draw("COLZ")
c.Draw()
c.SaveAs("figures/nsigma_tof_v_tpc_proton.pdf")
```
