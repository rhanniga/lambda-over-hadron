# Notebook for Correlations Meeting Presentation

```python
import ROOT as rt
```

```python
canvas = rt.TCanvas("canvas", "Main Canvas", 0, 10, 1600, 1200)

canvas.SetRightMargin(0.09);
canvas.SetLeftMargin(0.15);
canvas.SetBottomMargin(0.15);
```

```python
input_file = rt.TFile('trig_4_8_assoc_2_4_cent_0_20_mixcorr_hLambda.root')

hh2D = input_file.Get("hh2D")
uncorrhh2D = input_file.Get("uncorrhh2D")

hLambda2DLside = input_file.Get("hLambda2DLside")
hLambda2Dpeak = input_file.Get("hLambda2Dpeak")
hLambda2DRside = input_file.Get("hLambda2DRside")

uncorrhLambda2Dpeak = input_file.Get("uncorrhLambda2Dpeak")

rt.gStyle.SetOptStat(0)
```

```python
eta_max = 1.4
hh2D.GetXaxis().SetRangeUser(-eta_max, eta_max)
uncorrhh2D.GetXaxis().SetRangeUser(-eta_max, eta_max)
hLambda2Dpeak.GetXaxis().SetRangeUser(-eta_max, eta_max)
hLambda2DRside.GetXaxis().SetRangeUser(-eta_max, eta_max)
hLambda2DLside.GetXaxis().SetRangeUser(-eta_max, eta_max)
uncorrhLambda2Dpeak.GetXaxis().SetRangeUser(-eta_max, eta_max)

uncorrhh2D.SetTitle("Uncorrected h-h #Delta#varphi #Delta#eta distribution")
uncorrhh2D.GetXaxis().SetTitle("#Delta#eta")
uncorrhh2D.GetYaxis().SetTitle("#Delta#varphi")

hh2D.SetTitle("Mixed-event corrected h-h #Delta#varphi #Delta#eta distribution")
hh2D.GetXaxis().SetTitle("#Delta#eta")
hh2D.GetYaxis().SetTitle("#Delta#varphi")

uncorrhLambda2Dpeak.SetTitle("Uncorrected h-#Lambda^{0} #Delta#varphi #Delta#eta distribution")
uncorrhLambda2Dpeak.GetXaxis().SetTitle("#Delta#eta")
uncorrhLambda2Dpeak.GetYaxis().SetTitle("#Delta#varphi")

hLambda2Dpeak.SetTitle("Mixed-event corrected h-#Lambda^{0} #Delta#varphi #Delta#eta distribution")
hLambda2Dpeak.GetXaxis().SetTitle("#Delta#eta")
hLambda2Dpeak.GetYaxis().SetTitle("#Delta#varphi")
```

```python
hhDPhi = hh2D.ProjectionY().Clone("hhDPhi")
hLambdaDPhi = hLambda2Dpeak.ProjectionY().Clone("hLambdaDPhi")
hLambdaDPhiLside = hLambda2DLside.ProjectionY().Clone("hLambdaDPhiLside")
hLambdaDPhiRside = hLambda2DRside.ProjectionY().Clone("hLambdaDPhiRside")


hhDPhi.Sumw2()
hLambdaDPhi.Sumw2()

hhDPhi.SetTitle("h-h #Delta#varphi distribution")
hLambdaDPhi.SetTitle("h-#Lambda^{0} peak #Delta#varphi distribution")
hLambdaDPhiLside.SetTitle("h-#Lambda^{0} L-sideband #Delta#varphi distribution")
hLambdaDPhiRside.SetTitle("h-#Lambda^{0} R-sideband #Delta#varphi distribution")

hhDPhi.GetYaxis().SetTitle("dN_{h-h}/d#Delta#varphi")
hLambdaDPhi.GetYaxis().SetTitle("dN_{h-#Lambda^{0}}/d#Delta#varphi")
hLambdaDPhiLside.GetYaxis().SetTitle("dN_{h-#Lambda^{0}}/d#Delta#varphi")
hLambdaDPhiRside.GetYaxis().SetTitle("dN_{h-#Lambda^{0}}/d#Delta#varphi")


hhDPhi.SetLineColor(4)
hhDPhi.GetYaxis().SetRangeUser(1.25, 4)

hLambdaDPhi.SetLineColor(8)
hLambdaDPhiLside.SetLineColor(4)
hLambdaDPhiRside.SetLineColor(6)
hLambdaDPhi.GetYaxis().SetRangeUser(0.005, 0.0115)

hhDPhi.SetLineWidth(3)
hLambdaDPhi.SetLineWidth(3)
hLambdaDPhiRside.SetLineWidth(3)
hLambdaDPhiLside.SetLineWidth(3)

hhDPhi.SetMarkerColor(4)
hLambdaDPhi.SetMarkerColor(8)
hLambdaDPhiLside.SetMarkerColor(4)
hLambdaDPhiRside.SetMarkerColor(6)
```

```python
hhDPhi.Draw()
canvas.Draw()
```
