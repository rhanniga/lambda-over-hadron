```python
import ROOT as rt
```

```python
c = rt.TCanvas("c", "c", 800, 600)
tof_input_file = rt.TFile("/home/astrooo/Wonderland/native/research/lambda-over-hadron/data/offline/output/v0_avg6_rsb_116_118_sig_11_1134_trig_40_80_assoc_20_40_delta_eta_12.root")
input_file = rt.TFile("output/v0_avg6_sideband_subtraction_rsb_1135_115_sig_11_1134_trig_40_80_assoc_20_40_delta_eta_12_normal.root")

tof_dphi = tof_input_file.Get("h_lambda_dphi_subtracted_0_20")
dphi = input_file.Get("h_lambda_dphi_subtracted_0_20")

tof_dphi.Divide(dphi)

pol0 = rt.TF1("pol0", "[0]", -rt.TMath.Pi()/2, 3*rt.TMath.Pi()/2)

tof_dphi.Fit(pol0, "R")

tof_dphi.GetYaxis().SetRangeUser(0, 1)
tof_dphi.Draw()
c.Draw()
```
