```python
import ROOT as rt
rt.gStyle.SetOptStat(0)
```

```python
c = rt.TCanvas("c","c",800,600)
```

```python
infile = rt.TFile("eff_out_mult_eta_dep.root")

# Get the efficiency histograms
mb_assoc_pt_eff = infile.Get("fAssociatedEff")
mb_trigger_pt_eff = infile.Get("fTriggerEff")
mb_lambda_pt_eff = infile.Get("fLambdaEff")
mb_lambdaV0_pt_eff = infile.Get("fLambdaV0Eff")

mb_assoc_pt_eta_eff = infile.Get("fAssociatedPtEtaEff_mult_0_80")
mb_trigger_pt_eta_eff = infile.Get("fTriggerPtEtaEff_mult_0_80")
mb_lambda_pt_eta_eff = infile.Get("fLambdaPtEtaEff_mult_0_80")
mb_lambdaV0_pt_eta_eff = infile.Get("fLambdaV0PtEtaEff_mult_0_80")

lowmult_lambda_pt_eta_eff = infile.Get("fLambdaPtEtaEff_mult_0_20")
lowmult_lambdaV0_pt_eta_eff = infile.Get("fLambdaV0PtEtaEff_mult_0_20")

highmult_lambda_pt_eta_eff = infile.Get("fLambdaPtEtaEff_mult_50_80")
highmult_lambdaV0_pt_eta_eff = infile.Get("fLambdaV0PtEtaEff_mult_50_80")

mb_assoc_pt_eta_eff_comp = mb_assoc_pt_eta_eff.ProjectionX("mb_assoc_pt_eta_eff_comp")
mb_assoc_pt_eta_eff_comp.SetLineColor(rt.kGreen-2)
mb_assoc_pt_eta_eff_comp.SetMarkerColor(rt.kGreen-2)
mb_assoc_pt_eta_eff_comp.SetMarkerStyle(20)
mb_assoc_pt_eta_eff_comp.SetMarkerSize(1.5)
mb_assoc_pt_eta_eff_comp.Scale(mb_assoc_pt_eff.GetMinimum()/mb_assoc_pt_eta_eff_comp.GetMinimum())

mb_trigger_pt_eta_eff_comp = mb_trigger_pt_eta_eff.ProjectionX("mb_trigger_pt_eta_eff_comp")
mb_trigger_pt_eta_eff_comp.SetLineColor(rt.kRed-2)
mb_trigger_pt_eta_eff_comp.SetMarkerColor(rt.kRed-2)
mb_trigger_pt_eta_eff_comp.SetMarkerStyle(20)
mb_trigger_pt_eta_eff_comp.SetMarkerSize(1.5)
mb_trigger_pt_eta_eff_comp.Scale(mb_trigger_pt_eff.GetMinimum()/mb_trigger_pt_eta_eff_comp.GetMinimum())

mb_lambda_pt_eta_eff_comp = mb_lambda_pt_eta_eff.ProjectionX("mb_lambda_pt_eta_eff_comp")
mb_lambda_pt_eta_eff_comp.SetLineColor(rt.kBlue-2)
mb_lambda_pt_eta_eff_comp.SetMarkerColor(rt.kBlue-2)
mb_lambda_pt_eta_eff_comp.SetMarkerStyle(20)
mb_lambda_pt_eta_eff_comp.SetMarkerSize(1.5)
mb_lambda_pt_eta_eff_comp.Scale(mb_lambda_pt_eff.GetMinimum()/mb_lambda_pt_eta_eff_comp.GetMinimum())

mb_lambdaV0_pt_eta_eff_comp = mb_lambdaV0_pt_eta_eff.ProjectionX("mb_lambdaV0_pt_eta_eff_comp")
mb_lambdaV0_pt_eta_eff_comp.SetLineColor(rt.kBlack)
mb_lambdaV0_pt_eta_eff_comp.SetMarkerColor(rt.kBlack)
mb_lambdaV0_pt_eta_eff_comp.SetMarkerStyle(20)
mb_lambdaV0_pt_eta_eff_comp.SetMarkerSize(1.5)
mb_lambdaV0_pt_eta_eff_comp.Scale(mb_lambdaV0_pt_eff.GetMinimum()/mb_lambdaV0_pt_eta_eff_comp.GetMinimum())

lowmult_lambda_pt_eta_eff_comp = lowmult_lambda_pt_eta_eff.ProjectionX("lowmult_lambda_pt_eta_eff_comp")
lowmult_lambda_pt_eta_eff_comp.SetLineColor(rt.kMagenta-2)
lowmult_lambda_pt_eta_eff_comp.SetMarkerColor(rt.kMagenta-2)
lowmult_lambda_pt_eta_eff_comp.SetMarkerStyle(20)
lowmult_lambda_pt_eta_eff_comp.SetMarkerSize(1.5)

lowmult_lambdaV0_pt_eta_eff_comp = lowmult_lambdaV0_pt_eta_eff.ProjectionX("lowmult_lambdaV0_pt_eta_eff_comp")
lowmult_lambdaV0_pt_eta_eff_comp.SetLineColor(rt.kOrange-2)
lowmult_lambdaV0_pt_eta_eff_comp.SetMarkerColor(rt.kOrange-2)
lowmult_lambdaV0_pt_eta_eff_comp.SetMarkerStyle(20)
lowmult_lambdaV0_pt_eta_eff_comp.SetMarkerSize(1.5)

highmult_lambda_pt_eta_eff_comp = highmult_lambda_pt_eta_eff.ProjectionX("highmult_lambda_pt_eta_eff_comp")
highmult_lambda_pt_eta_eff_comp.SetLineColor(rt.kCyan-2)
highmult_lambda_pt_eta_eff_comp.SetMarkerColor(rt.kCyan-2)
highmult_lambda_pt_eta_eff_comp.SetMarkerStyle(20)
highmult_lambda_pt_eta_eff_comp.SetMarkerSize(1.5)

highmult_lambdaV0_pt_eta_eff_comp = highmult_lambdaV0_pt_eta_eff.ProjectionX("highmult_lambdaV0_pt_eta_eff_comp")
highmult_lambdaV0_pt_eta_eff_comp.SetLineColor(rt.kGray+1)
highmult_lambdaV0_pt_eta_eff_comp.SetMarkerColor(rt.kGray+1)
highmult_lambdaV0_pt_eta_eff_comp.SetMarkerStyle(20)
highmult_lambdaV0_pt_eta_eff_comp.SetMarkerSize(1.5)

```

```python
mb_assoc_pt_eff.Draw()
mb_assoc_pt_eta_eff_comp.Draw("same")
c.Draw()
```

```python
mb_trigger_pt_eff.Draw()
mb_trigger_pt_eta_eff_comp.Draw("same")
c.Draw()
```

```python
mb_lambda_pt_eff.Draw()
mb_lambda_pt_eta_eff_comp.Draw("same")
c.Draw()
```

```python
mb_lambdaV0_pt_eff.Draw()
mb_lambdaV0_pt_eta_eff_comp.Draw("same")
c.Draw()
```
