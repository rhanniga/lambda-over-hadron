```python
import ROOT as rt
rt.gStyle.SetOptStat(0)
```

```python
c = rt.TCanvas('c','c', 800, 600)
c.SetLeftMargin(0.15)
```

```python
ifile_newnorm = rt.TFile.Open("h_lambda_mc_closure_newnorm.root")
ifile_oldnorm = rt.TFile.Open("h_lambda_mc_closure_oldnorm.root")
newnorm_dist = ifile_newnorm.Get("h_lambda_dphi_subtracted_newnorm")
newnorm_dist_mc = ifile_newnorm.Get("h_lambda_dphi_subtracted_mc_newnorm")
oldnorm_dist = ifile_oldnorm.Get("h_lambda_dphi_subtracted_oldnorm")
oldnorm_dist_mc = ifile_oldnorm.Get("h_lambda_dphi_subtracted_mc_oldnorm")
```

```python
newnorm_dist.Draw()
newnorm_dist.GetYaxis().SetRangeUser(0.002, 0.0085)
oldnorm_dist.SetLineColor(rt.kBlue - 2)
oldnorm_dist.SetMarkerColor(rt.kBlue - 2)
oldnorm_dist.Draw("SAME")
leg = rt.TLegend(0.6, 0.7, 0.9, 0.85)
leg.SetBorderSize(0)
leg.AddEntry(newnorm_dist, "Norm. to B(0, [#pi/2, 3#pi/2])", "lep")
leg.AddEntry(oldnorm_dist, "Norm. to B(0, 0)", "lep")
leg.Draw()
# newnorm_dist.Divide(oldnorm_dist)
c.Draw()
c.SaveAs("different_norm_comp.pdf")
```

```python
ratio = newnorm_dist.Clone("ratio")
ratio.Divide(oldnorm_dist)
ratio.GetYaxis().SetTitle("B(0, [#pi/2, 3#pi/2]) / B(0, 0)")
ratio.GetYaxis().SetRangeUser(0.6, 1.4)
ratio.Draw()
c.Draw()
c.SaveAs("different_norm_comp_ratio.pdf")
```
