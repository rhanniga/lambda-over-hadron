```python
import ROOT as rt
```

```python
canvas = rt.TCanvas("main_canvas")
```

```python
infile = rt.TFile("../online/output/cent_0_90.root")
inlist = infile.Get("h-lambda_eff")
```

```python
real_lambda = inlist.FindObject("fRealNoDecayCutLambdaDist")
reco_lambda_trackCuts = inlist.FindObject("fTrackRecoLambdaDist")
```

```python
real_lambda.GetAxis(0).SetRangeUser(0.5, 10)
reco_lambda_trackCuts.GetAxis(0).SetRangeUser(0.5, 10)
real_lambda.GetAxis(2).SetRangeUser(-0.8, 0.8)
reco_lambda_trackCuts.GetAxis(2).SetRangeUser(-0.8, 0.8)
real_lambda.GetAxis(4).SetRangeUser(-10, 10)
reco_lambda_trackCuts.GetAxis(4).SetRangeUser(-10, 10)
real_lambda.GetAxis(5).SetRangeUser(1.11, 1.12)
reco_lambda_trackCuts.GetAxis(5).SetRangeUser(1.11, 1.12)
```

```python
real_lambda_pt = real_lambda.Projection(0)
reco_lambda_trackCuts_pt = reco_lambda_trackCuts.Projection(0)
```

```python
reco_lambda_trackCuts_pt.Integral()/real_lambda_pt.Integral()
```

```python
real_lambda_pt.Draw()
canvas.Draw()
```

```python
reco_lambda_trackCuts_pt.Draw()
canvas.Draw()
```
