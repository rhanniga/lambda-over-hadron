```python
import ROOT as rt
```

```python
in_file = rt.TFile("AnalysisResults.root")
in_list = in_file.Get("h-lambda_eff")
```

```python
c = rt.TCanvas("main", "Main Canvas", 0, 10, 1920, 1080)
```

```python
px_diff = in_list.FindObject("fPxDifference")
px_diff_fb = in_list.FindObject("fPxDifferenceFB")
py_diff = in_list.FindObject("fPyDifference")
pz_diff = in_list.FindObject("fPzDifference")
pt_diff = in_list.FindObject("fPtDifference")
```

```python
px_fit.Delete()
```

```python
px_diff.SetTitle("p_{x}^{real}-p_{x}^{reco} for tracks that pass daughter cuts")
px_fit = rt.TF1("px_fit", "[0]*TMath::Voigt(x - [1], [2], [3], 4)")
px_fit.SetNpx(1000)
px_fit.SetParameter(0, 1000)
px_fit.SetParameter(1, 0)
px_fit.SetParameter(2, 0.01)
px_fit.SetParameter(3, 0.01)
px_diff.Sumw2()
px_diff.GetXaxis().SetRangeUser(-0.1, 0.1)
# px_diff.GetYaxis().SetRangeUser(0, 0)
px_diff.SetLineColor(rt.kRed)
# px_diff.Fit(px_fit)
px_diff.Draw()
c.Draw()
```

```python
px_diff_fb.SetTitle("p_{x}^{real}-p_{x}^{reco} for tracks with 1024 filter bit")
px_fit_fb = rt.TF1("px_fit_fb", "gaus")
px_fit_fb.SetNpx(1000)
px_fit_fb.SetParameter(0, 80000)
px_fit_fb.SetParameter(1, 0)
px_fit_fb.SetParameter(2, 0.02)
px_diff_fb.Sumw2()
px_diff_fb.GetXaxis().SetRangeUser(-0.1, 0.1)
px_diff_fb.SetLineColor(rt.kRed)
px_diff_fb.Fit(px_fit_fb)
px_diff_fb.Draw()
c.Draw()
```

```python
py_diff.SetTitle("p_{y}^{real}-p_{y}^{reco} for tracks that pass daughter cuts")
py_diff.Sumw2()
py_diff.GetXaxis().SetRangeUser(-0.1, 0.1)
py_diff.SetLineColor(rt.kGreen + 2)
py_diff.Draw()
c.Draw()
```

```python
pz_diff.SetTitle("p_{z}^{real}-p_{z}^{reco} for tracks that pass daughter cuts")
pz_diff.Sumw2()
pz_diff.GetXaxis().SetRangeUser(-0.1, 0.1)
pz_diff.SetLineColor(rt.kBlue)
pz_diff.Draw()
c.Draw()
```

```python
pt_diff.SetTitle("p_{i}^{real}-p_{i}^{reco} for tracks that pass daughter cuts")
pt_diff.Sumw2()
pt_diff.GetXaxis().SetRangeUser(-0.1, 0.1)
pt_diff.SetLineColor(rt.kMagenta)
pt_diff.GetYaxis().SetRangeUser(0, 100000)
pt_diff.Draw()
px_diff.Draw("SAME")
py_diff.Draw("SAME")
pz_diff.Draw("SAME")
c.Draw()
```
